# An Efficiency Test of Sorting Algorithms

```python
import matplotlib.pyplot as pp
```

### 50 Random Numbers:

```python
Algorithm=["Insertion", "Merge", "Quick", "Quick I", "Quick II", "Quick III"]
CompileTime=[0.52, 0.55, 0, 2.05, 0, 0]
pp.bar(Algorithm, CompileTime, color=['red','blue','yellow','purple', 'green', 'orange'])
ax=pp.axes()
ax.set_facecolor("black")
pp.xlabel("Sorting Algorithms")
pp.ylabel("Compile Time")
pp.show()
```
